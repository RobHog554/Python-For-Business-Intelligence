### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.

````python
# Task 1

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]


# a.
# Valid transactions for each product require valid values for price (strictly positive) and quantity sold (non-negative)

valid_transactions = []
for transaction in transactions:
    if transaction["price"] > 0 and transaction["quantity"] >= 0:
        valid_transactions.append(transaction)


# b. Define a new field in the same way as you would access any other value through a key in a dict
for transaction in valid_transactions:
    transaction["total"] = transaction["price"] * transaction["quantity"]


# c. For every valid transaction, get the revenue from the region of the valid transaction (if the key exists, otherwise revenue = 0 for this transaction) and add the current revenue to the revenue stored in totalrev_by_region
totalrev_by_region = {}
for transaction in valid_transactions:
    totalrev_by_region[transaction["region"]] = totalrev_by_region.get(transaction["region"], 0) + transaction["total"]
    # Example of how the line works for each region
    # The first line does the following: totalrev_by_region["EU"] = 0 + 2400
    # The fifth line does the following: totalrev_by_region["EU"] = 2400 + 1200
print(totalrev_by_region)


# d. Grab largest value of TR
top_rev = max(totalrev_by_region.values())

# Grab the key to the max TR by region (i.e. the region with the largest max TR) 
top_region = max(totalrev_by_region, key=totalrev_by_region.get) 

print(f"The region with the largest total revenue is {top_region}, with a total of {top_rev}.")

````

### Task 2: Creating a function (1p)


You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
# Task 2

def summarise_sales(daily_sales):
    sales_metrics = {}

    sales_metrics["total"] = sum(daily_sales)
    sales_metrics["average"] = round(sum(daily_sales)/len(daily_sales), 2)
    sales_metrics["best_day"] = daily_sales.index(max(daily_sales)) + 1
    sales_metrics["worst_day"] = daily_sales.index(min(daily_sales)) + 1

    # Add 1 to an implicit counter variable if the sales figure of a given day is above the average
    sales_metrics["above_average_count"] = sum(1 for sale in daily_sales if sale > sales_metrics["average"])
    
    return sales_metrics


# Daily sales in EUR over 14 days 
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
print(summarise_sales(sales))
    
````

Using `for` loops and `if` statements, perform the following:

b) Loop through the students and print a status message for each. If the student's grade is 80 or above, print `"[name]: Excellent"`; if between 50 and 79, print `"[name]: Pass"`; otherwise print `"[name]: Fail"`.

c) Count how many students have an attendance rate below 0.80. Print the count.

d) Find the student with the highest grade using a loop (do not use `max()`). Print the student's name and grade.

e) Now find the student with the highest grade using `max()`. Print the student's name and grade.

````python
# Task 3

students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a. 
passing = []
for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])
print(passing)
print()


# b.
for s in students:
    if s["grade"] >= 80:
        s_status = "Excellent"
        
    elif s["grade"] >= 50:
        s_status = "Pass"

    else: s_status = "Fail"
    print(f"{s['name']}: {s_status}")
print()


# c.
low_attendance = sum(1 for student in students if student["attendance"] < 0.80)
print(f"Amount of students with a low attendance rate: {low_attendance}")


# c. Using loops instead of list comprehension
low_attendance_count = 0
for s in students:
    if s["attendance"] < 0.80:
        low_attendance_count += 1
print(f"{low_attendance_count} students have low attendance rate")
print()


# d.
top_grade = 0
top_student = ""
for s in students:  
    if top_grade < s["grade"]: 
        top_grade = s["grade"]
        top_student = s["name"] 
    
print(f"{top_student} got a {top_grade}, the highest grade in the class") 
        

# e. Take max value of every object in the list (students) that is obtained using the a certain key (grade)
best_student = max(students, key=lambda student: student["grade"])
print(f"{best_student['name']}: {best_student['grade']}")
````

### Task 4: Class and methods (1p)

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
# Task 4

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a. 
    # Nested dictionaries; ticker is key for outer while shares and price_per_share are keys for the inner
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}
        
    # b.
    def total_value(self):
        mv_of_assets = 0
        # Use .values to access the inner dict to get shares and price per share for each asset/ stock
        for asset in self.holdings.values(): 
            mv_of_assets += asset["shares"] * asset["price_per_share"]
        return mv_of_assets

    # c.
    def most_valuable(self):
        top_mv = 0
        top_mv_ticker = ""
        # Ticker is "AAPL" and asset is {"shares": 10, "price_per_share": 150}. Use .items to unpack both ticker and values in inner dict
        for ticker, asset in self.holdings.items():
            mv_of_asset = asset["shares"] * asset["price_per_share"]
            if mv_of_asset > top_mv:
                top_mv = mv_of_asset
                top_mv_ticker = ticker
        return top_mv_ticker # The question asked to only return the ticker
        
    
    # d.
    def __str__(self):
    # Summarize holdings; return owner name, number of assets and total value
        return f"Owner: {self.owner}\nNumber of assets in portfolio: {len(self.holdings)}\nTotal market value of portfolio: {self.total_value()}"

        


# Create an instance to test functionality of Portfolio
portfolio = Portfolio("Teuvo Teknolog")
assets = [
    ("AAPL", 10, 150.0),
    ("TSLA", 5, 200.0),
    ("GOOGL", 8, 100.0),
    ]

# Add data for all variables 
for ticker, shares, price in assets:
    portfolio.add_asset(ticker, shares, price)

print(portfolio.total_value())
print(portfolio.most_valuable())
print(portfolio)
````
