### Exercise 5: Creating a DataFrame

a) Create a DataFrame called `products` from the following dictionary:

```python

    "product": ["Laptop", "Phone", "Tablet", "Monitor"],

    "units_sold": [150, 300, 200, 100]

```

b) Print the DataFrame's `index` and `columns`.

c) Add a new column called `revenue` that is the product of `price` and `units_sold`.

````python
import pandas as pd

product_data = {
    "product": ["Laptop", "Phone", "Tablet", "Monitor"],
    "price": [1200, 800, 450, 350],
    "units_sold": [150, 300, 200, 100]
}

products = pd.DataFrame(product_data)

print(products.index)
print(products.columns)

products["revenue"] = products["price"] * products["units_sold"]

print(products)
````

### Exercise 10: Putting It All Together

A small shop tracks weekly sales in a dictionary:

```python
sales_data = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "customers": [30, 45, 50, 35, 60, 80, 70],
    "sales_eur": [600, 900, 1100, 750, 1300, 1800, 1500]
}
```

a) Create a DataFrame from this dictionary.

b) Add a column `avg_spend` that computes the average spending per customer (i.e., `sales_eur / customers`).

c) Using boolean indexing, filter the DataFrame to show only days where the number of customers exceeded 50.

d) Using `.iloc`, select the data for the weekend (last two rows) and all columns except `day`.

e) What was the total revenue for the week? Use NumPy or Pandas to compute and print the answer.

````python
# YOUR CODE HERE

sales_data = pd.DataFrame ({
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "customers": [30, 45, 50, 35, 60, 80, 70],
    "sales_eur": [600, 900, 1100, 750, 1300, 1800, 1500]
})

sales_data["avg_spend"] = round(sales_data["sales_eur"] / sales_data["customers"], 2)

print(sales_data[sales_data["customers"] > 50])
print()

print(sales_data.iloc[-2:, 1:4])
print()


total_revenue = sales_data["sales_eur"].sum()
print(f"Total revenue of the whole week is {total_revenue} euros")
````
